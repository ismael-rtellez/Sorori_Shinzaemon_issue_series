# Problem 1: Number of grains of rice on the 100th day

```python
grain = 1
grain_per_day = [] #Create a list for storing rice grains
total_grain_per_day = [] #Create a list for storing total rice grains
i = 1
grain_per_day.append(grain)       #Add the first
total_grain_per_day.append(grain) #value to both lists
while (i < 100):    #We use a while to repeat the operation until we reach 100 days
  i += 1            #Increment the counter
  grain = grain * 2 #Multiply the rice grains
  grain_per_day.append(grain) #Add value to the list
  total_grain_per_day.append(total_grain_per_day[-1] + grain) #We add with the last value and add to the list

print("Number of grains of rice on the 100th day: {}".format(grain_per_day[-1]))
print("The total rice grains after 100 days: {}".format(total_grain_per_day[-1]))
```

```python
# Graphical representation
import matplotlib.pyplot as plt
#We customize the graph of the amount of rice grains per day
fig, ax = plt.subplots()
ax.plot(grain_per_day, color="green", linestyle=":", linewidth=3, label="Daily Rice Grains")
plt.title("Daily Rice Grains over 100 Days")
ax.set_xlabel("Days")
ax.set_ylabel("Number of grains")
ax.set_title("Number of grains of rice per day")
plt.legend()
plt.show()

#We customize the graph of the total amount of rice grains up to the 100th day
fig, ax = plt.subplots()
ax.plot(total_grain_per_day, color="red", linestyle="--", linewidth=3, label="Total Daily Rice Grains")
plt.title("Total Daily Rice Grains over 100 Days")
ax.set_xlabel("Days")
ax.set_ylabel("Number of grains")
ax.set_title("Number of grains of rice per day")
plt.legend()
plt.show()
```

# Problem 2: Number of rices grains outside of the 100th day

```python
def compute_sorori_shinzaemon(n_days=100):
    "A function that calculates the number of rice grains per day in the rice anecdote of Shinzaemon Sorori."
    list_n_grains = []      #Create
    list_total_grains = []  #two list
    grain = 1

    for day in range(n_days): #We use a for to repeat the operation and calculate the total number of rice grains
      if day == 0:
        list_n_grains.append(grain)     #Add the first
        list_total_grains.append(grain) #value the both list
      else:
        grain = grain * 2   #Multiply the rice grains
        list_n_grains.append(grain)   #Add value to the list
        list_total_grains.append(list_total_grains[-1] + grain) #We add with the last value and add to the list

    return list_n_grains, list_total_grains

n_days = 32 #We assign the number of days for which we are going to perform the calculation
list_n_grains, list_total_grains = compute_sorori_shinzaemon(n_days=n_days) #we call the function with the required parameters
print(f"Number of grains of rice on the {n_days} day: {list_n_grains[-1]}")
print(f"The total rice grains after {n_days} days: {list_total_grains[-1]}")

#graphical representation
#Customize the graphic as desired
#We plot both the list of rice grains per day and the total rice grains at the last day
fig, ax = plt.subplots()
ax.plot(list_n_grains, color="blue", linewidth=2, label="Daily Rice Grains", marker="x", markersize=6)
ax.plot(list_total_grains, color="red", linewidth=2, label="Tatal Daily Rice Grains", marker="o", markersize=6)
plt.title(f"Daily Rice Grains over {n_days} Days")
ax.set_xlabel("Days")
ax.set_ylabel("Number of grains")
ax.set_title("Number of grains of rice per day")
plt.legend()
plt.show()
```

# Problem 3: How many people can live for how many days with the rice we get?

```python
def compute_days_of_rice(number_of_rice_grains, number_of_people):
  "Calculate how many days a given number of peoplecan live of the available rice grains"


  "Using data from the navigator, let's say that the daily consumption of rice for one person is 150 grams"
  "With the weight of each grain of rice we can calculate how many grains are equivalent to 150g"
  daily_rice_grain_person = 6000  #150 grams / 0.025 grams (weight of each rice grain)

  total_daily_consumption = daily_rice_grain_person * number_of_people  #We calculate the daily consumption of all people

  number_of_days_to_live = number_of_rice_grains //total_daily_consumption  #We calculate the number of days you can live

  return int(number_of_days_to_live)

number_of_rice_grains = list_total_grains[-1] #We assign the number of rice grains according to the last
                                              #value of the total list of grains in the previous problem

number_of_people = 5    #We assign the number of people for whom the calculation will be made
days_people_can_live = compute_days_of_rice(number_of_rice_grains, number_of_people)  #we call the function with the required parameters
print("With {:,.0f} grains of rice, {} people can live {:,.0f} days.".format(number_of_rice_grains, number_of_people, days_people_can_live))
```
